```python
import numpy as np
import pandas as pd
```

```python
# Import Data
breweries = pd.read_csv('/Users/josephyuenpro/Documents/WPI/CS573/a4/breweries_us.csv')
```

```python
# Drop duplicates
breweries = pd.DataFrame.drop_duplicates(breweries)

# Filter out unecessary headers - might add in later for an additional feature
breweries_filtered = breweries.filter(["brewery_name", "type", "state", "state_breweries"])

# Eliminate Puerto Rico as it is not a US state
breweries_filtered = breweries_filtered[breweries_filtered['state'] != 'puerto-rico']
```

```python
# Replace names of headers that are double counts
# DECISION - made closed breweries count still as a part of their brewery type, so Brewpub-Closed is counted as a Brewpub
breweries_filtered.loc[breweries_filtered.type == 'Contract', 'type'] = 'ContractBrewery'
breweries_filtered.loc[breweries_filtered.type == 'Mircobrewery', 'type'] = 'Microbrewery'
breweries_filtered.loc[breweries_filtered.type == 'Brewpub-Closed', 'type'] = 'Brewpub'
breweries_filtered.loc[breweries_filtered.type == 'Microbrewery-Closed', 'type'] = 'Microbrewery'
breweries_filtered.loc[breweries_filtered.type == 'BOP-BrewOnPremise-Closed', 'type'] = 'BOP-BrewOnPremise'
breweries_filtered.loc[breweries_filtered.type == 'ContractBrewery-Closed', 'type'] = 'ContractBrewery'
```

```python
brewery_types = breweries_filtered.type.unique()
brewery_types
```

```python
for brewery_type in brewery_types:
    temporary_df = breweries_filtered[breweries_filtered['type'] == brewery_type]
    temporary_df_group = temporary_df.groupby('state')
    temporary_df_group = pd.DataFrame(temporary_df_group.size().reset_index(name = 'state_' + brewery_type))
    breweries_filtered = pd.merge(breweries_filtered, temporary_df_group, how='left', on='state')
    breweries_filtered['state_' + brewery_type] = breweries_filtered['state_' + brewery_type].fillna(0)
```

```python
breweries_filtered['state'] = breweries_filtered['state'].str.capitalize()
```

```python
breweries_filtered.to_csv('breweries_cleaned.csv', index = False)
```
